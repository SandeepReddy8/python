```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_file = 'product_stock.csv'  
data = pd.read_csv('product_stock.csv')
```

```python
data.head()
```

```python
product_names = data['Product Name']
stock_values = data['Stock']
```

```python
plt.figure(figsize=(8, 8))
plt.pie(stock_values, labels=product_names, autopct='%1.1f%%', startangle=140)
plt.title("Product Stocks Available")


plt.show()
```
